**Q1. What is Object-Oriented Programming (OOP)?**

**Q2. What is a class in OOP?**

Ans2- In Object-Oriented Programming (OOP), a class serves as a blueprint or template for creating objects. It is a user-defined data type that defines the structure and behavior that its objects will possess.

Ans3- In Object-Oriented Programming (OOP), an object is a fundamental building block that represents a real-world entity or concept within a software system. It is an instance of a class, which acts as a blueprint defining the object's structure and behavior.

**Q4. What is the difference between abstraction and encapsulation?**

Ans4- Abstraction is the concept of hiding the complex implementation details and showing only the essential features of an object, while encapsulation is the practice of bundling data and methods together in a single unit and restricting direct access to the data to protect it.

**Q5. What are dunder methods in Python?**

Ans5- Dunder methods, also known as magic methods or special methods, are a core feature of Python that enable objects to interact with built-in functions, operators, and language constructs in a customized way. **Q6. Explain the concept of inheritance in OOP?**

**Q7. What is polymorphism in OOP?**

**Q9. What is a constructor in Python?**

**Q10. What are class and static methods in Python?**

Ans10- In Python, both class methods and static methods are defined within a class but differ in their binding and access to class or instance state.
Class Methods:

Decorator: Defined using the @classmethod decorator.
First Parameter: Takes cls as its first parameter, which refers to the class itself. This allows the method to access and modify class attributes and call other class methods.

Static Methods:

Decorator: Defined using the @staticmethod decorator.
First Parameter: Does not take self (instance) or cls (class) as an implicit first parameter. It behaves like a regular function, but is encapsulated within the class's namespace.

**Q17. What is the difference between a class variable and an instance variable?**

Ans17- The fundamental difference between instance variables and class variables lies in their scope, ownership, and how they are accessed and stored.

Instance Variables:

Ownership: Belong to a specific instance (object) of a class. Each object created from a class will have its own independent copy of instance variables.

Declaration: Declared within a class but outside of any method, constructor, or block, and without the static keyword (in languages like Java).

Lifetime: Created when an object is instantiated and destroyed when the object is garbage-collected.

Access: Accessed through an object reference (e.g., objectName.instanceVariable).

Purpose: Used to store data that is unique to each individual object.
Class Variables (Static Variables):

Ownership: Belong to the class itself, not to any specific instance. All objects of that class share a single copy of the class variable.

Declaration: Declared within a class but outside of any method, constructor, or block, and with the static keyword (in languages like Java).

Lifetime: Created when the class is loaded into memory and exists for the entire duration the program runs or until the class is unloaded.

Access: Can be accessed using the class name (e.g., ClassName.classVariable) or through an object reference.

Purpose: Used to store data that is common to all objects of a class, or to represent properties of the class itself.

**Q18. What is multiple inheritance in Python?**

Ans18- Multiple inheritance in Python allows a class to inherit from more than one parent class, enabling it to acquire attributes and methods from all of its base classes. This differs from single inheritance, where a class inherits from only one parent.

**Q19. Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python?**

Ans19- In Python, __str__ and __repr__ are special methods (often called "dunder methods" due to their double underscores) that define how an object is represented as a string. They serve different purposes and are intended for different audiences:

__str__ (for users):
Purpose: To provide an informal, human-readable, and concise string representation of an object. It's meant for end-users and should be easily understandable.

__repr__ (for developers):
Purpose: To provide a formal, unambiguous, and detailed string representation of an object, primarily for developers and debugging. The goal is that, ideally, eval(repr(obj)) should reconstruct an equivalent object.

**Q20. What is the significance of the ‘super()’ function in Python?**

Ans20- The super() function in Python holds significant importance within the context of object-oriented programming, particularly concerning inheritance. Its primary significance lies in facilitating the interaction between a child class and its parent (or super) class.

**Q21.  What is the significance of the __del__ method in Python?**

Ans21- The __del__ method in Python, also known as the destructor, holds significance primarily in resource management and cleanup operations when an object is about to be destroyed.

**Q22. What is the difference between @staticmethod and @classmethod in Python?**

Ans22- The primary difference between @staticmethod and @classmethod in Python lies in their access to class and instance data.

@classmethod:
Takes cls as the first argument: A class method receives the class itself as its first argument, conventionally named cls. This allows the method to access and modify class-level attributes and call other class methods.

@staticmethod:
No implicit first argument: A static method does not receive self (the instance) or cls (the class) as its first argument. It behaves like a regular function defined within the class's namespace.

**Q23.  How does polymorphism work in Python with inheritance?**

Ans23- Polymorphism in Python, when combined with inheritance, allows objects of different classes to be treated as objects of a common type, enabling a single interface to be used for various implementations. This is primarily achieved through method overriding.

**Q24. What is method chaining in Python OOP?**

Ans24- Method chaining in Python OOP is a programming technique that allows for calling multiple methods sequentially on the same object in a single expression. This creates a concise and readable flow of operations, often referred to as a "fluent API."

**Q25. What is the purpose of the __call__ method in Python?**

Ans25- The __call__ method in Python serves to make instances of a class callable, allowing them to be invoked like functions. When an object with a defined __call__ method is called using parentheses (e.g., my_object()), the __call__ method of that object is executed.

```python
# Q4. . Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes
#Sparrow and Penguin that override the fly() method.

class Bird:
    def fly(self):
        print("Birds can fly")

class Sparrow:
    def fly(self):
        print("Sparrows can fly at high speeds")

class Penguin:
    def fly(self):
        print("Penguins cannot fly")

bird = Bird()
bird.fly()

sparrow = Sparrow()
sparrow.fly()

penguin = Penguin()
penguin.fly()

obj = [bird, sparrow, penguin]
def parcer(obj):
  for i in obj:
    i.fly()
parcer(obj)
```

```python
# Q5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
#balance and methods to deposit, withdraw, and check balance.

class Bank:
    def __init__(self, balance):
        self._balance = balance
    def deposit(self, amount):
        self._balance += amount
    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
        else:
            print("Insufficient balance")
    def check_balance(self):
        return self._balance

acc1 = Bank(1000)
acc1.deposit(500)
acc1.withdraw(200)
print(acc1.check_balance())
```

```python
# Q9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the
#fraction as "numerator/denominator"?

class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Example usage:
fraction1 = Fraction(3, 4)
print(fraction1)
```

```python
# Q10.  Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
#vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"
```

```python
vector1 = Vector(2, 3)
vector2 = Vector(5, 7)

vector3 = vector1 + vector2
print(vector3)
```

```python
# Q12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute
#the average of the grades.

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        total = sum(self.grades)
        average = total / len(self.grades)
        return average
```

```python
student1 = Student("Alice", [85, 90, 78, 92])
average = student1.average_grade()
print(f"{student1.name}'s average grade is: {average}")
```

```python
# Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
#and hourly rate. Create a derived class Manager that adds a bonus to the salary?

class Employee:
    def __init__(self, name, hourly_rate):
        self.name = name
        self.hourly_rate = hourly_rate

    def calculate_salary(self, hours_worked):
        return hours_worked * self.hourly_rate

class Manager(Employee):
    def __init__(self, name, hourly_rate, bonus):
        super().__init__(name, hourly_rate)
        self.bonus = bonus

emp = Employee("Naved", 15)
salary = emp.calculate_salary(40)
print(f"{emp.name}'s salary: ${salary:.2f}")
```

```python
# Q15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that
#calculates the total price of the product.

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

prod = Product("Laptop", 1000, 2)
print(f"Total price: ${prod.total_price()}")
```

```python
# Q18. Create a class House with attributes address and price. Create a derived class Mansion that adds an
#attribute number_of_rooms.

class House:
  def __init__(self, address, price):
    self.address = address
    self.price = price

class Mansion(House):
  def __init__(self, address, price, number_of_rooms):
    super().__init__(address, price)
    self.number_of_rooms = number_of_rooms

  def house_info(self):
    print(f"Address: {self.address}")
    print(f"Price: {self.price}")
    print(f"Number of rooms: {self.number_of_rooms}")


mansion = Mansion("123 Main St", 250000, 5)
mansion.house_info()
```
